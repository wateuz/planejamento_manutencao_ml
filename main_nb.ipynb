```python
# Import de tabelas excel para iniciar o processo de ETL

import pandas as pd

df = pd.read_excel('4_Dados_PI_LIC_22311020.xlsx')
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format="mixed")
# Convertendo a coluna 'Value' para numrica, forando erros a NaN
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')
df.dropna(subset=['Value'], inplace=True)
df['value_sp'] = 0
df['value_mv'] = 0
df = df.rename(columns={'Value': 'value_pv', 'Timestamp': 'timestamp'})
dados_em_lote = df[['value_pv', 'timestamp', 'value_sp', 'value_mv']].to_dict(orient='records')
dados_em_lote
```

```python
# Cadastro em banco de dados (sqlite)
import pandas as pd
import datetime
from sqlalchemy.orm import sessionmaker
from models.sensores_db import LeituraSensor, engine, Base

# Garante a criao do banco de dados
Base.metadata.create_all(engine)

# Estabelece a sesso
Session = sessionmaker(bind=engine)
session = Session()

session.bulk_insert_mappings(LeituraSensor, dados_em_lote)
session.commit()
resultados = session.query(LeituraSensor).all()
for resultado in resultados:
    print(resultado)
session.close()
```
