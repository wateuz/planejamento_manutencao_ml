```python
# Recorrencia do texto breve e padronização do texto breve
import pandas as pd
import re
import unicodedata

# 'df' é DataFrame com a coluna 'TEXTO_BREVE'

# 1. Regex para extrair o TAG
regex_tag = r'(p[v|ic]{1,2}-?\d+[\w-]*[a-z]?)'

# 2. Extrai o TAG para uma nova coluna (case-insensitive)
df['TAG_EQUIPAMENTO_mod'] = df['TEXTO_BREVE'].str.extract(regex_tag, flags=re.IGNORECASE, expand=False).str.upper()

# 3. Cria uma coluna de descrição SEM o TAG (case-insensitive)
df['TEXTO_BREVE_SEM_TAG'] = df['TEXTO_BREVE'].str.replace(regex_tag, ' ', flags=re.IGNORECASE, regex=True)


# 4. Sua função de padronização (perfeita como está)
def padronizar_texto(texto):
    if not isinstance(texto, str):
        return ''
    texto = texto.lower().strip()
    texto = ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')
    texto = re.sub(r'[^\w\s]', '', texto)
    stop_words = set(['a', 'o', 'e', 'de', 'da', 'do', 'para', 'pela', 'por', 'no', 'na'])
    palavras = [palavra for palavra in texto.split() if palavra not in stop_words]
    texto = ' '.join(palavras)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto


# 5. Aplica a função de padronização na coluna correta (a que já está sem o TAG)
df['TEXTO_BREVE_PADRONIZADO'] = df['TEXTO_BREVE_SEM_TAG'].apply(padronizar_texto)


# 6. Faz a contagem final na coluna completamente limpa
contagem_final = df['TEXTO_BREVE_PADRONIZADO'].value_counts()

print(contagem_final.head(20))
df
```

```python
import datetime
df_result['DATA_INICIO_REAL'] = pd.to_datetime(df_result['DATA_INICIO_REAL'], utc=True, format="mixed")
df_result['DATA_FIM_REAL'] = pd.to_datetime(df_result['DATA_FIM_REAL'], utc=True, format="mixed")
df_result_asc = df_result.sort_values(by='DATA_INICIO_REAL', ascending=True).reset_index(drop=True)
# Exemplo: itera diariamente entre min_serie_temporal e max_serie_temporal
df_temporal_manutenção = pd.DataFrame()
for i in df_result_asc.index:
    print(i)
    start_date = df_result_asc.loc[i, 'DATA_INICIO_REAL']
    finish_date = df_result_asc.loc[i, 'DATA_FIM_REAL']
    print(start_date.date(),' - ' , finish_date.date())
    if pd.isna(start_date) or pd.isna(finish_date):
        print("AVISO: A data de início ou de fim é inválida (NaT). O processamento será pulado.")
    else:
        for j in pd.date_range(min_serie_temporal, max_serie_temporal, freq='D'):
            print(j)
            if start_date <= j <= finish_date:
                df_temporal_manutenção = pd.concat([df_temporal_manutenção, pd.DataFrame({'DATA_EM_MAN': [j], 'Value': [0]})], ignore_index=True)
            elif start_date > j <= finish_date:
                df_temporal_manutenção = pd.concat([df_temporal_manutenção, pd.DataFrame({'DATA_EM_MAN': [j], 'Value': [1]})], ignore_index=True)

df_temporal_manutenção
```
